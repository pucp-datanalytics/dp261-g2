# 01_business  
## Proyecto: Predicción de Cancelaciones Hoteleras

Este cuaderno presenta la definición del problema de negocio para el proyecto de analítica enfocado en la predicción de cancelaciones de reservas hoteleras.

Se abordan los siguientes componentes:

- Problema de negocio  
- Variable objetivo  
- KPIs  
- Criterios de éxito

### 1. Problema de negocio
En el sector hotelero, las cancelaciones de reservas representan un problema relevante porque generan pérdidas económicas, reducen la ocupación real y dificultan la planificación operativa. Cuando una reserva se cancela, el hotel puede perder oportunidades de venta, afectar su proyección de ingresos y tener menor capacidad para gestionar de manera eficiente sus habitaciones disponibles.

Actualmente, la gestión de cancelaciones suele ser reactiva: el hotel actúa una vez que la cancelación ya ocurrió, en lugar de anticiparse a ella. Esto limita la posibilidad de aplicar estrategias preventivas como confirmaciones anticipadas, políticas de depósito más adecuadas, promociones dirigidas o sobreventa controlada.

## Definición del problema:

El hotel no cuenta con un sistema predictivo que permita identificar con anticipación qué reservas tienen alta probabilidad de cancelarse, lo que afecta negativamente los ingresos, la ocupación y la toma de decisiones operativas.

## Propósito del análisis:

Desarrollar un modelo predictivo que permita anticipar cancelaciones de reservas y apoyar decisiones orientadas a reducir pérdidas y optimizar la ocupación hotelera.

### 2. Variable objetivo
La variable objetivo del proyecto es:

  **Nombre:** is_canceled

**Tipo:** Binaria

  **Interpretación:**

    1: la reserva fue cancelada

    0: la reserva no fue cancelada

**Tipo de problema analítico:**

    Se trata de un problema de clasificación binaria supervisada, ya que el objetivo es predecir si una reserva será cancelada o no a partir de variables históricas asociadas a la reserva y al cliente.


## 3. KPIs
Para evaluar el valor del proyecto, se proponen indicadores en tres niveles: negocio, operación y desempeño del modelo.

 **3.1 KPIs de negocio**
    Tasa de cancelación (%)
    Proporción de reservas canceladas respecto al total de reservas.

 - **Revenue perdido por cancelaciones**
    Monto estimado de ingresos no concretados debido a cancelaciones.

 - **No-show rate**
    Porcentaje de clientes que no se presentan sin completar adecuadamente la reserva.

**3.2 KPIs operativos**
- **Ocupación efectiva**
    Porcentaje real de habitaciones ocupadas.

- **Ocupación proyectada vs. ocupación real**
    Diferencia entre la ocupación esperada y la finalmente ejecutada.

- **Overbooking controlado**
    Nivel de sobreventa aplicado estratégicamente para compensar cancelaciones esperadas.

**3.3 KPIs del modelo predictivo**
- **Accuracy**
    Proporción total de predicciones correctas.

- **Precision**
    Proporción de reservas predichas como canceladas que realmente se cancelan.

- **Recall**
    Capacidad del modelo para identificar correctamente las reservas que sí se cancelan.

- **F1-score**
    Medida balanceada entre precision y recall.

- **AUC-ROC**
    Capacidad general del modelo para discriminar entre reservas canceladas y no canceladas.

    Nota: En este problema, el recall es especialmente importante, porque interesa detectar la mayor cantidad posible de cancelaciones reales para intervenir a tiempo.

### 4. Criterios de éxito
El éxito del proyecto debe evaluarse tanto por el rendimiento técnico del modelo como por su utilidad práctica para el negocio.

**4.1 Criterios de éxito del modelo**
   
    - Alcanzar un recall ≥ 80% en la clase de cancelación.
    
    - Lograr un AUC ≥ 0.75.
    
    - Superar el desempeño de un modelo base o una regla simple de negocio.

**4.2 Criterios de éxito del negocio**
    - Reducir la tasa de cancelación efectiva en al menos un 10%.
    - Incrementar la ocupación efectiva.
    - Disminuir las pérdidas asociadas a cancelaciones y no-show.

**4.3 Criterios de éxito operativo**
El modelo debe ser útil para activar medidas concretas, como:

    - Confirmaciones anticipadas de reservas.
   
    - Políticas diferenciadas de depósito o prepago.
   
    - Campañas de retención o descuentos dirigidos.
   
    - Estrategias de overbooking basadas en riesgo estimado.
